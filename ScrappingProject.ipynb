```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
URL="https://www.webscraper.io/test-sites/e-commerce/static/computers/laptops"
page=requests.get(URL)
soup=BeautifulSoup(page.content,"html.parser")
```

```python
soup
```

```python
titre=soup.find_all("a" , class_="title")
description=soup.find_all("p" , class_="description card-text")
prix=soup.find_all("h4" , class_="float-end price card-title pull-right")
titre
```

```python
# Je dois d√©finir tous les divisions du premier Laptop pour ne pas pas travailler chaque division toute seule

Job_elements=soup.find_all("div",class_="card-body")
Job_elements
```

```python
# class c'est sur l'inspection sur le site web
# "p", "a", "h4" c'est des balises, on les trouve en inspectant les √©l√©ments

data_list=[]
for i in range(0,
len(Job_elements)):
  description = Job_elements[i].find("p",class_="description card-text").get_text(strip=True)
  titre = Job_elements[i].find("a" , class_="title").text
  prix = Job_elements[i].find("h4" , class_="float-end price card-title pull-right").get_text(strip=True)
  data_list.append((titre,description,prix))
```

```python
# Afficher le Data_list
data_list
```

```python
# Pour l'instant j'ai scrapper les donn√©es que d'une seule page

# Afficher le DataFrame
df = pd.DataFrame(data_list,columns=['titre', 'Description', 'prix'])
df
```

```python
# Pour l'instant je vais scrapper les donn√©es pour les 20 pages

# Je vais int√©grer une nouvelle boucle dans l'ancienne boucle

data_list=[]
for j in range(0,20):
  URL = f"https://www.webscraper.io/test-sites/e-commerce/static/computers/laptops?page={j}"

  # J'ai red√©fini l'URL car j'ai prix un autre

  page=requests.get(URL)
  soup=BeautifulSoup(page.content,"html.parser")

  #J'ai red√©fini le Job_elements faite que j'ai pris un nouveau URL

  Job_elements=soup.find_all("div",class_="card-body")
  for i in range(0, len(Job_elements)):
     description = Job_elements[i].find("p",class_="description card-text").get_text(strip=True)
     titre = Job_elements[i].find("a" , class_="title").text
     prix = Job_elements[i].find("h4" , class_="float-end price card-title pull-right").get_text(strip=True)
     data_list.append((titre,description,prix))
```

```python
# Afficher le nouveau DataFrame

df = pd.DataFrame(data_list,columns=['titre', 'Description', 'prix'])
df
```
